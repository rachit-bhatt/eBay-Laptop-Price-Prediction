```python
import pandas as pd
import re
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import random
from selenium.webdriver.common.keys import Keys  # Import the Keys module

# Initialize Chrome driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

df = []
count=0



pages = ["https://www.ebay.com/b/Apple-Laptops/111422/bn_320025?_pgn={}".format(page) for page in range(130, 210)]

for page_url in pages:
    driver.get(page_url)
    





    
    elements = driver.find_elements(By.CLASS_NAME, 's-item__link')

    for element in elements:
        try:

            href = element.get_attribute('href')
 
            
            driver.execute_script("window.open('', '_blank');")
            driver.switch_to.window(driver.window_handles[-1])
            driver.get(href)  # Visit the individual listing page

            title_element = driver.find_element(By.CLASS_NAME, 'x-item-title__mainTitle')
            title = title_element.text

            price_element = driver.find_element(By.CLASS_NAME, 'x-price-primary')
            price = price_element.text

            try:
                ram_element = driver.find_element(By.CLASS_NAME, 'ux-labels-values--ramSize')
                ram = ram_element.find_element(By.CLASS_NAME, 'ux-labels-values__values-content').text
            except:
                ram = 'N/A'

            try:
                brand_element = driver.find_element(By.CLASS_NAME, 'ux-labels-values--brand')
                brand = brand_element.find_element(By.CLASS_NAME, 'ux-labels-values__values-content').text
            except:
                brand = 'N/A'
                
            try:
                condtion_element = driver.find_element(By.CLASS_NAME, 'ux-labels-values--condition')
                condition = condtion_element.find_element(By.CLASS_NAME, 'ux-labels-values__values-content').text
            except:
                condition = 'N/A'
            try:
                harddrive_element = driver.find_element(By.CLASS_NAME, 'ux-labels-values--hardDriveCapacity')
                harddrive = harddrive_element.find_element(By.CLASS_NAME, 'ux-labels-values__values-content').text
            except:
                harddrive = 'N/A'
            try:
                ssd_element = driver.find_element(By.CLASS_NAME, 'ux-labels-values--ssdCapacity')
                SSD = ssd_element.find_element(By.CLASS_NAME, 'ux-labels-values__values-content').text
            except:
                SSD = 'N/A'
            try:
                Snotes_element = driver.find_element(By.CLASS_NAME, 'ux-labels-values--sellerNotes')
                notes = Snotes_element.find_element(By.CLASS_NAME, 'ux-labels-values__values-content').text
            except:
                notes = 'N/A'
                
            try:
                rating_elements = driver.find_elements(By.CLASS_NAME, 'fdbk-detail-seller-rating__value')
                ratings = [element.text for element in rating_elements]
            except:
                ratings='N/A'
            
            try:
                review_elements = driver.find_elements(By.CLASS_NAME, 'fdbk-container__details__comment')
                reviews = [element.text for element in review_elements]
            except:
                reviews='N/A'
                
                
            
            df.append({"href":href,"Title": title, "Price": price, "RAM": ram, "Brand": brand,"Hdd":harddrive,"Condition":condition,"Seller notes":notes,"SSD":SSD,"Rating":ratings,"Reviews":reviews})
            
            
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            
            
            time.sleep(random.uniform(1, 3))
        except Exception as e:
            print(f"An error occurred: {e}")
            count+=1

driver.quit()


print(df)
```

```python

ssd_pattern = r'(\d+(?:GB|TB)\sSSD)'

def extract_ssd(title):
    ssd_match = re.search(ssd_pattern, title)
    if ssd_match:
        return ssd_match.group()
    else:
        return None


missing_ssd_indices = data4[data4['SSD'].isnull()].index
data4.loc[missing_ssd_indices, 'SSD'] = data4.loc[missing_ssd_indices, 'Title'].apply(extract_ssd)
```

```python

ssd_pattern = r'(\d+GB\s(?:RAM|DDR\d+))'

def extract_ssd(title):
    ssd_match = re.search(ssd_pattern, title)
    if ssd_match:
        return ssd_match.group()
    else:
        return None


missing_ssd_indices = data4[data4['RAM'].isnull()].index
data4.loc[missing_ssd_indices, 'RAM'] = data4.loc[missing_ssd_indices, 'Title'].apply(extract_ssd)
```

```python
def mode_safe(x):
    mode_val = x.mode()
    if len(mode_val) > 0:
        return mode_val.iloc[0]
    else:
        return np.nan

```

```python

dataset['SSD'] = dataset.groupby('Brand')['SSD'].transform(lambda x: x.fillna(mode_safe(x)))

dataset['Hdd'] = dataset.groupby('Brand')['Hdd'].transform(lambda x: x.fillna(mode_safe(x)))

dataset['RAM'] = dataset.groupby('Brand')['RAM'].transform(lambda x: x.fillna(mode_safe(x)))
```

## creatign the columns for rating

```python
def parse_ratings(ratings):
    categories = ['Accurate description', 'Reasonable shipping cost', 'Shipping speed', 'Communication']
    rating_values = {}
    
    for i, category in enumerate(categories):
        rating_values[category] = ratings[i] if len(ratings) > i else None
    
    return rating_values


dataset[['Accurate description', 'Reasonable shipping cost', 'Shipping speed', 'Communication']] = dataset['Rating'].apply(parse_ratings).apply(pd.Series)
```

## NLP

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def filter_reviews(reviews_data: str) -> str:
    reviews_tokens = word_tokenize(reviews_data.lower())
    english_stopwords = stopwords.words('english')
    filtered_reviews_tokens = [review_token for review_token in reviews_tokens if review_token not in english_stopwords]
    lemmatizer = WordNetLemmatizer()
    lemmatized_reviews_tokens = [lemmatizer.lemmatize(review_token) for review_token in filtered_reviews_tokens]
    lemmatized_reviews = ' '.join(lemmatized_reviews_tokens)
    filtered_reviews = re.findall(pattern=r'\w+|\d+', string=lemmatized_reviews)
    return ' '.join(filtered_reviews)

dataset['filtered_reviews'] = dataset['Reviews'].apply(lambda x: filter_reviews(x))

print(dataset)
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

def is_review_positive(review: str) -> bool:
    sentiment_intensity_analyzer = SentimentIntensityAnalyzer()
    review_sentiment = sentiment_intensity_analyzer.polarity_scores(review)
    if review_sentiment['compound'] >= 0.05:
        return 1  # Positive
    elif review_sentiment['compound'] <= -0.05:
        return -1  # Negative
    else:
        return 0  # Neutral

dataset['review'] = dataset['filtered_reviews'].apply(is_review_positive)

print(dataset)
```
